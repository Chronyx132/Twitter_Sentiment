```python
import seaborn as sns
import matplotlib.pyplot as plt

accuracies = [
    82.27, 82.08, 81.78, 81.885, 48.85, 82.1, 82.41, 82.235,
    77.215, 50.175, 76.475, 82.225, 82.05, 82.17, 81.355,
    81.955, 82.04, 82.145, 81.685, 82.5, 82.38
]


# Second set of accuracies
accuracies2 = [
    81.99, 82.235, 82.09, 82.14, 49.75, 82.275, 82.355, 82.095,
    77.39, 74.95, 77.19, 82.09, 81.925, 82.045, 81.545,
    81.95, 81.78, 82.445, 82.205, 82.44, 82.235
]

# Create the horizontal box plot with two datasets
plt.figure(figsize=(10, 4))
plt.boxplot([accuracies, accuracies2], vert=False, patch_artist=True,
            labels=["Model 1", "Model 2"])

# Add labels and title
plt.xlabel("Accuracy (%)")
plt.title("Comparison of Accuracies: Model 1 vs Model 2")

# Display the plot
plt.tight_layout()
plt.show()
```

```python
# Create horizontal box plot without showing outliers
plt.figure(figsize=(10, 4))
plt.boxplot([accuracies, accuracies2], vert=False, patch_artist=True,
            labels=["Model 1", "Model 2"], showfliers=False)

# Add labels and title
plt.xlabel("Accuracy (%)")
plt.title("Comparison of Accuracies (Outliers Removed): Model 1 vs Model 2")

# Display the plot
plt.tight_layout()
plt.show()
```
